```python
filename = "data/sample2condition.txt"
```

```python
with open(filename, 'r') as f:
    data = f.readlines()
    samples = []
    for line in data:
        line = line.strip().split()
        samples.append(line)
        print(line)
```

```python
import pandas as pd # import! 
```

```python
df = pd.DataFrame(samples, columns = ['sample name', 'condition'])
df
```

```python
condition_counts = df['condition'].value_counts().reset_index().rename(columns =
    {'condition':'count', 'index':'condition type'}
    )
condition_counts
```

```python
import seaborn as sns # import!
import matplotlib.pyplot as plt # import!
```

```python
sns.barplot(data = condition_counts, x = 'condition type', y = 'count')
plt.xticks(rotation = 90)
plt.show()
```
